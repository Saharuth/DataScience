```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('https://github.com/Saharuth/DataScience/raw/main/disney_plus_titles.csv')
df.head()
```

```python
df.hist()
```

```python
df.release_year.hist()
```

ไม่ต้องการแสดง <Axes: > ให้ใส่ semicolon ปิดท้าย

```python
df.release_year.hist();
```

กำหนดจำนวนแท่งหรือ bins

```python
df.release_year.hist(bins=20)
```

ต้องการเอาเส้นกริดในกราฟออก ให้ใส่ grid=False

```python
df.release_year.hist(bins=20, grid=False, color='orange')
```

ถ้าต้องการแสดงกราฟมากกว่า 1 => df[['ชื่อคอลัมน์ที่ 1','ชื่อคอลัมน์ที่ 2']].hist()
โดยสามารถกำหนดให้แกน x แกน y มีสเกลที่เท่ากันได้ด้วย => sharex=True, sharey=True

ถ้าต้องการให้กราฟที่มีมากกว่า 2 แสดงอยู่ในพื้นที่กราฟเดียวกัน => df[['ชื่อคอลัมน์ที่ 1','ชื่อคอลัมน์ที่ 2']].plot.hist();
โดยถ้าต้องการปรับให้สีกราฟดูง่ายขึ้น => alpha=.2 ตัวเลขเพิ่มขึ้น สียิ่งเข้ม

```python
#density plot
df.release_year.plot.density()
```
